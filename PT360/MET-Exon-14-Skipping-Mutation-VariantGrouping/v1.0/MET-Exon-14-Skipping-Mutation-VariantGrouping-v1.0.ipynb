```python
#Filter applied: biomarker name = MET, test result name = poitive, equivocal and defined variant patterns
filtered_data = data[
    (data['biomarker_name_name'] == "MET") & 
    (data['test_result_name'].isin(['Positive', 'Equivocal'])) &
    (
        data['molecular_variant'].str.contains(keyword1, case=False, na=False) |
        data['molecular_variant'].str.contains(keyword2, case=False, na=False) |
        data['molecular_variant'].apply(lambda x: any(re.search(pattern, str(x)) for pattern in variant_patterns))
    )
]

#total molecular variants post filter for which variant grouping is required

filtered_data.groupby('molecular_variant').size().reset_index(name='count')
```

```python
#Check total sum
filtered_data.groupby('molecular_variant').size().reset_index(name='count').sum()
```

```python
# assign variant group MET exon 14 skipping to column variant_grouping_code, variant_grouping_name, and variant_grouping_vocab based on filtered data

filtered_data['variant_grouping_name'] = 'MET Exon 14 Skipping Mutation'

filtered_data['variant_grouping_code'] = 'C131179'

filtered_data['variant_grouping_vocab'] = 'NCIt'

#filtered_data.head()
```

```python
#Check variant grouping sum
filtered_data.groupby(['variant_grouping_name', 'variant_grouping_code', 'variant_grouping_vocab']).size().reset_index(name='count')
```

```python
#apply changes to the csv file

data.update(filtered_data)
```
